```python
import random
import pandas as pd
import numpy as np
import datetime 
from datetime import date
from datetime import datetime
import matplotlib 
import matplotlib.pyplot as plt
```

```python
def histogram_place(name):
    plt.figure(figsize=(16,12))
    data = place[name]
    data.dropna(how="all")
    print(data.describe())
    x = data.tolist()
    n, bins, patches = plt.hist(x, 
                                bins='auto',
                                edgecolor="#6A9662", 
                                color="#DDFFDD", 
                                alpha=0.9,  
                                rwidth=0.85)
    plt.grid()
    if name == "Place's time" or name == "Place's type":
        plt.xticks(rotation = 90)
    plt.title("Histogram of " + str(name))
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show() 
```

```python
def bar_place(namecol1, namecol2):
    print(place[[namecol1, namecol2]].describe())
    plt.figure(figsize=(16,12))
    plt.bar(place[namecol1], height= place[namecol2])
    plt.grid()
    plt.xticks(rotation = 90)
    plt.xlabel("Frequency")
    plt.ylabel("Values")
    plt.title("Barchart of "+ namecol2)
    plt.show()
```

```python
def histogram_review(name):
    plt.figure(figsize=(16,12))
    data = review[name]
    data.dropna(how="all")
    print(data.describe())
    x = data.tolist()
    plt.hist(x, bins='auto', color='#0504aa', alpha=0.9, rwidth=0.85)
    plt.grid()
    if name == "Reviewer's place":
        plt.xticks(rotation = 90)
    plt.title("Histogram of " + str(name))
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.show() 
```

```python
def bar_chart(col_name):
    plt.figure(figsize=(16,12))
    n = review.groupby(col_name)[col_name].count().head(len(review[col_name]))
    plt.bar(x= n.index, height= n)
    plt.grid()
    plt.xticks(rotation = 90)
    plt.xlabel("Frequency")
    plt.ylabel("Values")
    plt.title("Barchart of "+ str(col_name))
    plt.show()
```

```python
def pie_chart_place(name_col):
    data = place.groupby(name_col)[name_col].count().head(len(place[name_col]))
    labels = data.index
    fig, ax = plt.subplots(figsize=(16,12))
    ax.pie(data.values, labels = labels, autopct='%.0f%%')
    ax.set_title('Pie chart of '+ name_col)
    plt.show()
```

```python
def pie_chart_review(name_col):
    data = review.groupby(name_col)[name_col].count().head(len(review[name_col]))
    labels = data.index
    fig, ax = plt.subplots(figsize=(16,12))
    ax.pie(data.values, labels = labels, autopct='%.0f%%')
    ax.set_title('Pie chart of '+ name_col)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

N = 1000  # Number of samples

# Create synthetic dataset
X1 = np.random.normal(loc=0, scale=1, size=(N, 2))
Y1 = 0 * np.ones(shape=(1000,))  # LABEL = 0 

X2 = np.random.normal(loc=[-5, 5], scale=1, size=(N, 2))
Y2 = 1 * np.ones(shape=(1000,))  # LABEL = 1

X3 = np.random.normal(loc=[5, -5], scale=1, size=(N, 2))
Y3 = 2 * np.ones(shape=(1000,))  # LABEL = 2

# Create stacked dataset
X = np.vstack((X1, X2, X3))
Y = np.hstack((Y1, Y2, Y3))

# TRAIN SVM LEARNING ALGORITHM
clf = SVC(kernel='linear')
clf = clf.fit(X, Y)

# create decision boundary plot
xx, yy = np.meshgrid(
    np.arange(-10, 10, 0.2),
    np.arange(-10, 10, 0.2))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# PLOT EVERYTHING
plt.scatter(X1[:,0], X1[:,1], color='r')
plt.scatter(X2[:,0], X2[:,1], color='b')
plt.scatter(X3[:,0], X3[:,1], color='y')
plt.contourf(xx,yy,Z,cmap=plt.cm.coolwarm, alpha=0.8)
plt.title("SVM With Linear Kernel and Three Labels (0, 1, 2)")
plt.show()
```
